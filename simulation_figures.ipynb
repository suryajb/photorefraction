# Examples

instructions:
- set up pump modes and sh modes
- set up laser modes
- set up solver (solve_twm is the solver class, with pump laser inputs, and the corresponding ir and visible modes). Note that the order matters, it must be in the format of solve_twm([laser1,laser2,...],[pump1,pump2,...,sh_mode1,sh_mode2,...])
- run solver (main function is pr_shg_transmission())

```python
from laser_modes_prsolve import *
```

```python
%matplotlib inline
ir_pump_mode2 = cavity_mode(wavelength=1560e-9,Q0=5e5,Qc=5e5,vis=False,pump=True)
vis_sh_mode2 = cavity_mode(wavelength=780e-9,Q0=5e5,Qc=5e5,vis=True,pump=False)
pump_laser2 = laser(wavelength=1560e-9,start=1562.00e-9,end=1558e-9,scan_speed=0.000001e-9,sample_rate=0.001,
                   Pin=50e-6,modes=[ir_pump_mode2,vis_sh_mode2])
time_scan = solve_twm([pump_laser2],[ir_pump_mode2,vis_sh_mode2])
time_scan.pr_shg_transmission(fn1='test_scan_',fn2='1',
                             save=False,single_scan=True,many_plots=True)
```
